```python
def wordcloud_draw(dataset,color='black'):

    words = ' '.join(dataset)

    cleaned_word = ' '.join([word for word in words.split()

    if (word != 'news' and word != 'text')])

    wordcloud = WordCloud(stopwords = stop,

    background_color = color,

    width = 2500, height = 2500).generate(cleaned_word)

    plt.figure(1, figsize = (10,7))

    plt.imshow(wordcloud)

    plt.axis("off")

    plt.show()
```

```python
print("POLITICS Related Words:",end="")

wordcloud_draw(POLITICS)
```

```python
# convert in lowercase
df1['headline']=df1['headline'].apply(lambda x:" ".join(x.lower() for x in x.split()))
df1['headline'].head()
```

```python
#Remove punctuation marks
df1['headline']=df1['headline'].str.replace('[^\w\s]','')
df1['headline'].head()
```

```python
# removal of stopwords
df1['headline']=df1['headline'].apply(lambda x:" ".join(x for x in x.split() if x not in stop))
df1['headline'].head()
```

```python
#lemmatization
wordnet = WordNetLemmatizer()
df1["headline"].apply(lambda x: " ".join([wordnet.lemmatize(word)for word in x.split()]))
df1["headline"].head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score, classification_report,roc_curve,roc_auc_score
from sklearn.naive_bayes import GaussianNB
```

```python
X=df1["headline"].values
y=df1["category_id"].values
```

```python
X
```

```python
y
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vec1=TfidfVectorizer(max_features=2500,min_df=7,max_df=0.8)
vec_X=vec1.fit_transform(X).toarray()
```

```python
X_train,X_test,y_train,y_test=train_test_split(vec_X,y,test_size=0.2,random_state=0)
```

```python
print(vec_X.shape,X_train.shape,X_test.shape)
```

```python
svc = SVC()
   
```

```python
svc.fit(X_train,y_train)
```

```python
y_pred=svc.predict(X_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
model=RandomForestClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
y1_pred=model.predict(X_test)
```

```python
print(classification_report(y_test,y1_pred))
```

```python
model1=GaussianNB()
```

```python
model1.fit(X_train,y_train)
```

```python
y2_pred=model1.predict(X_test)
```

```python
print(classification_report(y_test,y2_pred))
```
